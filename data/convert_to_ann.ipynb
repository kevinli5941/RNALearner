```python
import pandas as pd
import h5py
import scanpy as sc
from scipy import sparse
```

```python
data0 = pd.read_hdf('recount3/mouse_0_1000_transpose.h5').T
data1 = pd.read_hdf('recount3/mouse_1001_5000_transpose.h5').T
data2 = pd.read_hdf('recount3/mouse_5001_8001_transpose.h5').T
data3 = pd.read_hdf('recount3/mouse_8002_10087_transpose.h5').T
data = pd.concat([data0, data1, data2, data3])
data
```

```python
adata = sc.AnnData(data, data.index.to_frame(), data.columns.to_frame())
adata.obs.columns = ['Sample ID']
adata.var.columns = ['Gene ID']
```

```python
adata.X = sparse.csr_matrix(adata.X)
```

```python
adata.write_h5ad("recount3/mouse_FULL_sparse.h5ad")
```

## Ensembl to Gene ID

```python
import pybiomart
import numpy as np
```

```python
dataset = pybiomart.Dataset(name='mmusculus_gene_ensembl', host='http://www.ensembl.org')
mappings = dataset.query(attributes=['ensembl_gene_id', 'external_gene_name'])
```

```python
adata1 = sc.read_h5ad("recount3/mouse_FULL_sparse.h5ad")
```

```python
adata1
```

```python
adata1.var.columns = ['Ensemble ID']
adata1.var
```

```python
def convert(e):
    x = mappings[mappings['Gene stable ID'] == e]
    if x.shape[0] != 0:
        return x['Gene name'].iloc[0]
    return np.NaN
gene_ids = adata1.var_names.str.slice(stop=18).map(convert)
gene_ids
```

```python
np.count_nonzero(~pd.isnull(gene_ids))
```

```python
adata1.var_names = gene_ids
```

```python
adata1.var
```

```python
project_id = adata1.X[:, -1].todense()
 
```

```python
projects = np.ravel(project_id) 
projects
```

```python
x=pd.DataFrame(adata1.obs["Sample ID"].copy())
x["project"] = projects.astype('int')
x
```

```python
adata1.obs = x
```

```python
np.NaN
```

```python
adata1.var.index = adata1.var_names.fillna("No GeneID") 
```

```python
adata1.var.index = adata1.var.index.str.upper()
adata1.var_names
```

```python
adata1.write_h5ad("recount3/mouse_FULL_sparse_with_geneIDs.h5ad")
```
